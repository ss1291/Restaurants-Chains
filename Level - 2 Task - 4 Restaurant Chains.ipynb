# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the XLSB file

```python
df = pd.read_excel('C:/Users/admin/Desktop/Cognifyz Internship Projects/dataset.xlsb', engine='pyxlsb')
df
```

```python
df.head(3)
```

# Identify restaurant chains

```python
restaurant_chains = df['Restaurant Name'].value_counts()

#Consider only names with more than one occurrence
restaurant_chains = restaurant_chains[restaurant_chains > 1]

print(f'Restaurant Chains :\n{restaurant_chains}')
```

# Group the dataset by restaurant chains and calculate their average ratings and popularity

```python
chain_ratings_popularity = df.groupby('Restaurant Name').agg({'Aggregate rating': 'mean', 'Votes': 'sum'})

# Filter for restaurant chains only
chain_ratings_popularity = chain_ratings_popularity[chain_ratings_popularity.index.isin(restaurant_chains.index)]

# Sort the chains by average rating in descending order
chain_ratings_popularity = chain_ratings_popularity.sort_values(by='Aggregate rating', ascending=False)

# Select top 10 restaurant chains
top_10_chains = chain_ratings_popularity.head(10)

print(top_10_chains)
```

# Plotting the average ratings and popularity of restaurant chains

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.barplot(x=top_10_chains.index, y=top_10_chains['Aggregate rating'], palette='Blues')
plt.xticks(rotation=45, ha='right')
plt.title('Average Rating of Top 10 Chains')

plt.subplot(1, 2, 2)
sns.barplot(x=top_10_chains.index, y=top_10_chains['Votes'], palette='Reds')
plt.xticks(rotation=45, ha='right')
plt.title('Popularity of Top 10 Chains')

plt.tight_layout()
plt.show()
```
